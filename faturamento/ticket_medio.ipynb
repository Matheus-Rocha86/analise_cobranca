```python
import matplotlib.pyplot as plt
from repositorio_de_banco import ConsultaExecutor
from conexao_banco_dados_resulth import ConexaoBancoDadosResulth
from queries_ import consultar_ticket_medio
```

```python
def criar_repositorio(
        data_inicial,
        data_final,
        conexao: ConexaoBancoDadosResulth,
) -> ConsultaExecutor:
    return ConsultaExecutor(
        data_inicial=data_inicial,
        data_final=data_final,
        conexao_banco=conexao
    )
```

```python
def main():
    data_inicial = input("Entre com a data inicial AAAA-MM-DD: ")
    data_final = input("Entre com a data final AAAA-MM-DD: ")

    repositorio = criar_repositorio(
        data_inicial=data_inicial,
        data_final=data_final,
        conexao=ConexaoBancoDadosResulth()
    )

    dados = repositorio._executar_consulta(query=consultar_ticket_medio())

    mes, total_faturado_mensal, total_quantidade_pre_venda, ticket_medio = zip(*dados)


    return {
        "mes": mes,
        "total_faturado_mensal": total_faturado_mensal,
        "total_quantidade_pre_venda": total_quantidade_pre_venda,
        "ticket_medio": ticket_medio
    }
```

```python
resultado_2024 = main()
```
